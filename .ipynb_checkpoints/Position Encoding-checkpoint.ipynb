```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (12., 4.)
```

We will visualize for 150 time steps with an 82-dimensional vector, since this is usually the size of a human pose representation

```python
T = 150
d = 82
```

### Sine Pose Encoding
So we follow the steps from "Attention is all you need", with the same frequency in their geometric progression, but we will visualize by only using sine, then cosine and finally the alternating encoding

```python
pose_encoding = np.zeros((T, d), dtype=np.float)
W = [1/(10000**(2*k/d)) for k in range(d)]
F = [np.sin for k in range(d)]
for t in range(T):
    for k, w, f in zip(range(d), W, F):
        pose_encoding[t, k] = f(w*t)
                
sns.heatmap(pose_encoding)
plt.title('Sin/Cos Position Encoding')
plt.xlabel('Position Encoding - d')
plt.ylabel('Time Step - t')
plt.show()
```

### Cosine Position Encoding

```python
pose_encoding = np.zeros((T, d), dtype=np.float)
W = [1/(10000**(2*k/d)) for k in range(d)]
F = [np.cos for k in range(d)]
for t in range(T):
    for k, w, f in zip(range(d), W, F):
        pose_encoding[t, k] = f(w*t)
                
sns.heatmap(pose_encoding)
plt.title('Cos Position Encoding')
plt.xlabel('Position Encoding - d')
plt.ylabel('Time Step - t')
plt.show()
```

###  Sin/Cos Position Encoding

```python
pose_encoding = np.zeros((T, d), dtype=np.float)
W = [1/(10000**(2*k/d)) for k in range(d)]
F = [np.sin if k%2==0 else np.cos for k in range(d)]
for t in range(T):
    for k, w, f in zip(range(d), W, F):
        pose_encoding[t, k] = f(w*t)
                
sns.heatmap(pose_encoding)
plt.title('Sin/Cos Position Encoding')
plt.xlabel('Position Encoding - d')
plt.ylabel('Time Step - t')
plt.show()
```

###  Tanh Position Encoding

```python
pose_encoding = np.zeros((T, d), dtype=np.float)
W = [1/(10000**(2*k/d)) for k in range(d)]
F = [np.tanh for k in range(d)]
for t in range(T):
    for k, w, f in zip(range(d), W, F):
        pose_encoding[t, k] = f(w*t)
                
sns.heatmap(pose_encoding)
plt.title('Sin/Cos Position Encoding')
plt.xlabel('Position Encoding - d')
plt.ylabel('Time Step - t')
plt.show()
```
